# Dow Jones Index Data Set 

В датасете представлены данные индекса Dow Jones для классификации ценных бумаг по наибольшей прибыли.

 - `quarter` - квартал(1 = Янв-Март; 2 = Апр-Июнь).
 - `stock` - аббревиатура ценной бумаги.
 - `date` - последний торговый день на недели.
 - `open` - цена ценной бумаги в начале недели.
 - `close` - цена ценной бумаги в конце недели.
 - `high` - самая высокая цена в течении недели.
 - `low` - самая изкая цена в течение недели.
 - `volume` - колличество ценных бумаг торгующихся в течение недели.
 - `percent_change_price` - процентное изменение цены на протяжение недели.
 - `percent_chagne_volume_over_last_wek` - процентное изменение количества торгующих ценных бумаг по сравнению с предыдущей неделью.
 - `previous_weeks_volume` количество ценных бумаг за предыдущую неделю.
 - `next_weeks_open` - цена открытия на следующей недели.
 - `next_weeks_close` - цена закрытия на следующей недели.
 - `percent_change_next_weeks_price` - процентное изенение цены на следующей недели.
 - `days_to_next_dividend` - дни до следующих дивидентов 
 - `percent_return_next_dividend` - процент вернувщихся деведентов в следующей выплате.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

##  KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 2)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

Осуществим классификацию и вычислим значение ошибки

```python
knn
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Другой способ подсчета ошибки

```python
err_test = 1 - knn.score(X_test, y_test)
err_test
```

С помощью confusion_matrix построим матрицу, которая позволяет визуализировать прогнозы классификатора. Элемент матрицы (i, j) равна числу объектов, которые, принадлежат группе i, но, по прогнозам, находятся в группе j.

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

# Подбор параметров

Найдем такое значение параметра k, при котором значение ошибки будет минимальным

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

Осуществим классификацию при k=1 и вычислим значение ошибки еще раз

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

лучшее значение ошибки при текущий параметрах.


Получилась очень большая ошибка на наших данных. Метод К средний плохо сработал.

Возможно это из-за того, что для данныго алгоритма у нас слишком много классов. У нас их 30. Попробуем уменьшить их количество до 5.

```python
data = pd.read_csv('dow_jones_index.data', engine = 'python')
a = data[data['stock'] == 'AA']
b = data[data['stock'] == 'XOM']
c = data[data['stock'] == 'MMM']
d = data[data['stock'] == 'BA']
e = data[data['stock'] == 'IBM']
data = pd.concat((a, b, c, d, e), axis = 0)
data = data.drop(['quarter','open','close','date', 'high', 'low', 
                  'next_weeks_open', 'next_weeks_close','days_to_next_dividend'], axis = 1)
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
data_describe = data.describe(include = [object])
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
y = data['stock']
data = data.drop(['stock'],axis=1)
data = data.fillna(data.median(axis = 0), axis = 0)
data_numerical = data[numerical_columns]
X = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train, y_train)
```

```python
knn
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
err_test = 1 - knn.score(X_test, y_test)
err_test
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

# Подбор параметров

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

В данном случаи получились более адекватные ошибки хоть и все равно далекие от того, что нужно.
